##1. Subgraph

A subgraph is a subset of the nodes of a network, and all of the edges linking these nodes. Any group of nodes can form a subgraph.

`Graph.subraph()` is used for when inducing subgraph on nodes, `Graph.edge_subgraph()` is on edges.

##2. Components

Component subgraphs (or simply components) are portions of the network that are disconnected from each other.

If the network is fully connected, `Graph.is_connected()` function returns `True`, otherwise, `False`.

##3. Clique [under construction]

A clique is defined as a maximal complete subgraph of a given graph—i.e., a group of people where everybody is connected directly to everyone else. The word “maximal” means that no other nodes can be added to the clique without making it less connected.
