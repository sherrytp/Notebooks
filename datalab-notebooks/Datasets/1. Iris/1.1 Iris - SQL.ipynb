### 2. Read using SQL

Delete the `iris` table if it exists. The `create` command will not create a table if it already exists.

```
%sql
drop table if exists iris
```

The `OK` output indicates that the command was successful.

Create the `iris` table from the datafile.

```
%sql
create temporary table iris 
using CSV 
options(path="/mnt/datalab-datasets/file-samples/iris.csv", 
        header=TRUE)
```

Note that the command succeeded.

Display all columns from the `iris` table.

```
%sql
select *
from iris
```

Note the `select` command in Spark SQL will display only the first 1000 rows of a table.
