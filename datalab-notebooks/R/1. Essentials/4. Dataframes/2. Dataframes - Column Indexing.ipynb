We will continue to look at the `iris` dataframe.

#### Retrieving Single Columns

There are (at least) three ways to retrieve a single column of a dataframe. 

But first, recall the columns of the `iris` dataframe.

```
%r
head(iris)
```

A single column can be retrieved by specifying inside a pair of square brackets and written after a comma:
- the column number
- the column name

An example of each:

```
%r
iris
```

```
%r
iris['Species', 'Seoal.Width']
```

A single column can be retrieved by placing the column name after a dollar sign (which follows the dataframe name.)

```
%r
iris$Species
```

Notice that, in all three cases, the single column is returned as a vector.

## Numeric Indexing

Numeric indexing is performed by writing a vector of column numbers within square brackets and to the right of the comma.

For example:

Recall that the columns of a dataframe are numbered from left to right. The first column is `Sepal.Length` and the fifth column is `Species`.

```
%r
head(iris)
```

```
%r
str(iris)
```

The column numbers to display are provided as a numeric vector inside square brackets and to the left of the comma.

For example:

```
%r
iris[,5:1]
```

```
%r
iris[,1:5]
```

Notice that the result is a dataframe, which makes sense.

## String Indexing

String indexing is performed by writing a vector of column names within square brackets and to the right of the comma.

For example,

```
%r
iris[,c('Species','Sepal.Width')]
```

Again, notice that the result is a dataframe (which makes sense.)

Logical indexing can be used to retrieve specific columns, but is not often used (and so not included here.)

__cbind()__ is a function that can be used to add columns to the dataframe. Let's test this function on a sample dataframe:

```
%r
dummydf=head(iris)
SerialNo=seq(from=100, to=105, by=1)
dummydf=cbind(SerialNo,dummydf)
dummydf
```
