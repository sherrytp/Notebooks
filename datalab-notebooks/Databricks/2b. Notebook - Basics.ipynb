## Attaching

The micro-cluster will terminate after 120 minutes of inactivity. In order to run commands in a notebook the micro-cluster must be running. If it is not running then it will need to be recreated and reattached to your notebook.

In order to make this recreating and reattaching processing automatic check the box by "_When running commands in Notebooks, automatically launch and attach to clusters without prompting_" in the "Notebook Settings" of "User Settings", which you can find by clicking on the person icon in the upper right corner of this page. 

Now the micro-cluster will be automatically recreated and the notebook will re-attach to the notebook, whenever you need it.

## Publishing

Notebooks can be shared (read only) with others over the web. To do so click "Publish" at the top of this page. Then copy the URL and click "Done."

Try it with this notebook and browse to the URL.

## Importing

Notice the "Import Notebook" button at the top right corner of the URL where you published this notebook. 

- Click that button and copy the URL 
- Open the "Workspace" and select "Import" from one of the _down arrow_ menus
- Select URL, paste the URL and select "Import"
