# Notebook - Markdown

![alt](http://www.scatter.com/images/DataLab_logo.jpg)

Notebooks contains cells, which visually look like rectangles. 

These cells contain: 
- code, which usually produces some output
- markdown, which produces formatted text. 

_This notebook will demonstrate how to write markdown cells to produce formatted text._

See the `/Workbooks/Databricks/0. Notebook - R Code` notebook for an introduction on creating and running cells that contain R code.

Click your mouse anywhere in this cell. Then hit the _Return/Enter_ key. Then hit the `ESC`key. Try this a few times until you get used to the difference between _edit mode_ (where you see markdown) and _command mode_ (where you see formatted text) for a markdown cell. 

Now edit this cell so that you can see the markdown code. 

Notice that
- paragraphs are preceded and followed by a blank line
- the first word in the cell is "`%md`", which indicates that the cell will contain markdown
- unnumbered list items are created from lines that start with a dash and a space
- text between two underscores is displayed in _italics_
- text between two pairs of underscores is displayed in __bold__ in _italics_
- There is noo spel checkr

Also notice that

1. numbered list items are created from lines that start with "1. ", "2. ", etc. 
1. it doesn't matter what number is used to at the beginning of the line

Use the hash mark `#` to create text in a larger/bold font to indicate a title or section names. For instance, 

# My Title

## My Section 1

### My Subsection 1a

### My Subsection 1b

## My Section 2

The cheat sheet for Databricks markdown is:
