# Databricks Components

The primary benefits Databricks provides are:
1. A __notebook interface__ in which you format text and run Python, R, Scala and Spark commands/programs
1. A __filesystem__ in which you can access uploaded files and external files, such as those in AWS S3
1. Ability to run __Spark__ programs from notebooks and as jobs
1. Easy configuration, creation and termination of __Spark clusters__ 

Each is described below.
