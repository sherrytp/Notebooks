## Orientation Data Charts 

```python
school = pd.read_excel("Orientation data 2019.xlsx", sheetname = "School", index_col = 0)
degree = pd.read_excel("Orientation data 2019.xlsx", sheetname = "Degree", index_col = 0)
fl_pt = pd.read_excel("Orientation data 2019.xlsx", sheetname = "Enrollment", index_col = 0)
session = pd.read_excel("Orientation data 2019.xlsx", sheetname = "EdSession")
```

```python
def bar_plot(df): 
    ind = np.arange(df.shape[0])    # The x locations for the groups 
    width = 0.35       # the width of the bars: can also be len(x) sequence 
    
    ta = df['total attendance']
    reg = df['Registered']
    att = df['attend']
    site = df['on site reg']
    
    fig, ax = plt.subplots()
    p1 = ax.bar(ind - width/2, att, width, label = "Registered Attendee")
    p2 = ax.bar(ind - width/2, site, width, bottom=att, label = "On-site Registration")
    
    p3 = ax.bar(ind + width/2, reg, width, label = "Registered")
    
    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xticks(ind)
    ax.set_xticklabels(df.index)
    ax.legend()
    fig.tight_layout()

    plt.show()

bar_plot(school)
bar_plot(degree)
bar_plot(fl_pt)
```

```python
plt.bar(np.arange(session.shape[0]), session['head count'], width = 0.5)
plt.ylabel("Head Count")
plt.title("Education Session Attendees")

# zip joins x and y coordinates in pairs
for x,y in zip(np.arange(session.shape[0]), session['head count']):

    plt.annotate(y, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

textstr = "\n 0: Life on the Heights, 1: Ed Tools, 2: Life in the City; \n 3: Live the Mission; 4: Return to the Academy; 5: Strive to Thrive; \n 6: How to Succeed in a Doc Program; 7: Financial Services"
plt.text(0.01, 1, textstr, fontsize=14, verticalalignment='top')
```
