```python
# How to groupby both the company name and the monthly date in one function 
#Two ways to do this : one is to create a new column "Monthly", another is to make the monthly data as the list index
# first, we create a column in data frame allstocks5yr['monthly']
allstocks5yr["monthly"] = [x[0:7] for x in allstocks5yr['date']]
```

```python
#check the df again to see the column 
allstocks5yr.head()
```

```python
#Groupby the data by name first then apply the lambda function to the monthly data 
dfallstocks5yr = allstocks5yr.groupby(['Name','monthly']).nth(0)
```

```python
dfallstocks5yr
```

```python
#To find the monthly return, use the close price and divide by the close prie of the month before minus 1 
dfallstocks5yr['return' ]= dfallstocks5yr['close'].shift(1)/dfallstocks5yr['close'] -1
```

```python
# Volume Change
dfallstocks5yr['volume_Change'] = dfallstocks5yr['volume'].shift(1)/dfallstocks5yr['volume']-1
```

```python
###
class pipeline:

    def process(self,df):
        df["monthly"] = [x[0:7] for x in df['date']]
        dfallstocks5yr = df.groupby(['Name','monthly']).nth(0)
        return dfallstocks5yr
    def add_feature(self,df):
        df['return' ]= df['close'].shift(1)/df['close'] -1
        df['volume_Change'] = df['volume'].shift(1)/df['volume']-1
        return df
    def __init__(self,df):
        df = self.process(df)
        df = self.add_feature(df)
        self.dfprocessed = df
```

```python
pipelined = pipeline(allstocks5yr)
```
